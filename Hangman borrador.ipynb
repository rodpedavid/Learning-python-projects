```python
import numpy as np
from random_word import RandomWords

r = RandomWords()
word = r.get_random_word()
len_word = len(word)

print("You are about to save the world...")
print("... or not")
print("You are asked to guess the word in order to do it")
print("Be careful! you just have 4 lives")
print("____________________________________")
print("So let's start...")
print("Guess the word to save the world x_x")
    
answer = list(word)

display = []
display.extend(answer)
used = []
used.extend(answer)

for i in range(len(display)):
        display[i] = "_"
print(' '.join(display))
print()

counter = 0
counter_life = 0 
    
while counter<len_word:
    guess = input("Try with a letter to save the world:")
    guess = guess.lower()
        
    for i in range(len_word):
        if answer[i] == guess and guess in used:
            display[i] = guess
            counter = counter+1
            used.remove(guess)
          
    if guess not in display:
        print("You're wrong, try harder, SAVE US!")
        counter_life= counter_life+1
        
        if counter_life >= 4:
            print("Oh weeee, you just get caught by THEM. Goodbye, soldier")
            break
    print(' '.join(display))    

if counter == len_word:
    print("Congratulations, you just saved the world. Thank you for your service, soldier :)") 
print("-----------------------------------------------------------------------")
print("-----------------------------------------------------------------------")
print("-----------------------------------------------------------------------")
print("Thank you for playing this little game, hope you enjoyed it :) have a nice day or evening? DRP")
```
